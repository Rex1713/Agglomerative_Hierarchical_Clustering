```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customer_data = pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
customer_data.head()
```

```python
customer_data.shape
```

```python
data = customer_data.iloc[:,3:5].values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(8,8))
plt.title("Customer Dendogram")
dend = shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
dend1=shc.dendrogram(shc.linkage(data,method='single'))
```

```python
dend1=shc.dendrogram(shc.linkage(data,method='complete'))
```

```python
dend1=shc.dendrogram(shc.linkage(data,method='average'))
```

```python
dend = shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering as ac
```

```python
cluster = ac(n_clusters = 5, affinity='euclidean',linkage='ward')
labels = cluster.fit_predict(data)
```

```python
labels
```

```python
plt.scatter(data[:,0], data[:,1], c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(data,labels)
```

```python
cluster = ac(n_clusters = 2, affinity='euclidean',linkage='ward')
labels = cluster.fit_predict(data)
```

```python
silhouette_score(data,labels)
```

```python
cluster = ac(n_clusters = 3, affinity='euclidean',linkage='ward')
labels = cluster.fit_predict(data)
```

```python
silhouette_score(data,labels)
```
